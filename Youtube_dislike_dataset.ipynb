```python
# Histogram:


for i in df.select_dtypes(include = "number"):
    sns.histplot(data = df, x = i)
    plt.show()

    
```

```python
# histogram is not looking good ==> plot kde plot

for i in df.select_dtypes(include = "number"):
    plt.figure(figsize=(15,6)),
    sns.kdeplot(data = df, x =i)
    plt.show()
```

```python
# Boxplot: ==> outllier detection

for i in df.select_dtypes(include = "number"):
    sns.boxplot(data = df, x = i)
    plt.show()
```

```python
# Bivariate analysis:

# scatter plot:

for i in ["view_count","likes", "comment_count"]:
    sns.scatterplot(data = df, x = i, y = "likes")
    plt.show()
```

```python
for i in ["view_count","likes", "comment_count"]:
    sns.scatterplot(data = df, x = i, y = "likes")
    plt.show()
```

```python
for i in df.select_dtypes(include = "number"):
    sns.lmplot(data = df, x=i, y = "likes")
    plt.show()
```

```python
# correlation :
# heatmap
 
co = df.select_dtypes(include = "number").corr()
for i in df.select_dtypes(include = "number"):
    sns.heatmap(data = co, annot = True)
    plt.show()
```

```python
df.dropna(inplace = True)    # will remove null values
df.isnull().sum()

# feature engginering;  ==> creating extra feature;

# convert published at ==> date with extra feature
df["date"] = pd.to_datetime(df["published_at"])

df.dtypes
```

```python
df["month"] = df["date"].dt.month_name()
df["month"]
```

```python
df["year"] = df["date"].dt.year
df["year"]
```
