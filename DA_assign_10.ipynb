```python
'''
1. Create grand alliances
a. 'INC','NCP', 'RJD', 'DMK', 'IUML',
'JMM','JD(s)','KC(M)','RLD','RSP','CMP(J)','KC(J)','PPI','MD' as UPA
b. 'BJP','SS', 'LJP', 'SAD', 'RLSP',
'AD','PMK','NPP','AINRC','NPF','RPI(A)','BPF','JD(U)','SDF','NDPP','MNF','RIDALOS','KM
DK','IJK','PNK','JSP','GJM','MGP','GFP','GVP','AJSU','IPFT','MPP','KPP','JKPC','KC(T)','BDJ
S','AGP','JSS','PPA','UDP','HSPDP','PSP','JRS','KVC','PNP','SBSP','KC(N)','PDF','MDPF' as
NDA
c. 'YSRCP','AAAP', 'IND', 'AIUDF', 'BLSP', 'JKPDP', 'JD(S)', 'INLD', 'CPI', 'AIMIM',
'KEC(M)','SWP', 'NPEP', 'JKN', 'AIFB', 'MUL', 'AUDF', 'BOPF', 'BVA', 'HJCBL',
'JVM','MDMK' as Others

'''
```

```python
# create grand alliances:
UPA = ['INC','NCP', 'RJD', 'DMK', 'IUML','JMM','JD(s)','KC(M)','RLD','RSP','CMP(J)','KC(J)','PPI','MD']
NDA = ['BJP','SS', 'LJP', 'SAD', 'RLSP','AD','PMK','NPP','AINRC','NPF','RPI(A)','BPF','JD(U)','SDF','NDPP','MNF','RIDALOS','KMDK','IJK','PNK','JSP','GJM','MGP','GFP','GVP','AJSU','IPFT','MPP','KPP','JKPC','KC(T)','BDJS','AGP','JSS','PPA','UDP','HSPDP','PSP','JRS','KVC','PNP','SBSP','KC(N)','PDF','MDPF']
Others = ['YSRCP','AAAP', 'IND', 'AIUDF', 'BLSP', 'JKPDP', 'JD(S)', 'INLD', 'CPI', 'AIMIM','KEC(M)','SWP', 'NPEP', 'JKN', 'AIFB', 'MUL', 'AUDF', 'BOPF', 'BVA', 'HJCBL', 'JVM','MDMK']
```

```python
# add alliences col to the 2009;

candidate2009["Alliences"] = candidate2009["Party Abbreviation"].apply(lambda x: "UPA" if x in UPA else "NDA" if x in NDA else "Others")
candidate2014["Alliences"] = candidate2014["Party Abbreviation"].apply(lambda x: "UPA" if x in UPA else "NDA" if x in NDA else "Others")

candidate2014
```

```python
# find out the counts for alliances:
```

```python
candidate2009["Alliences"].value_counts()
candidate2014["Alliences"].value_counts()
```

```python
# 2. Create Winning seats distribution by Major Political Parties & Alliances for 2009 & 2014

new_df_candidate2009 = candidate2009[candidate2009["Position"] == 1]
new_df_candidate2014 = candidate2014[candidate2014["Position"] == 1]
new_df_candidate2014
```

```python
# count plot for each alliances:

fig, ax = plt.subplots(figsize=(8, 6))
sns.countplot(x= new_df_candidate2009['Alliences'])
ax.bar_label(ax.containers[-1], fmt='Count:\n%.2f', label_type='center')
plt.title(" Seats distribution by Major Political Parties & Alliances for 2009")
plt.show()

fig,ax = plt.subplots(figsize = (8,6))
sns.countplot(x = new_df_candidate2014["Alliences"])
ax.bar_label(ax.containers[-1], fmt = "count: \n %.2f", label_type = "center")
plt.title(" Seats distribution by Major Political Parties & Alliances for 2014")
plt.show()
```

```python
# 4. Plot comparatively seats won based on candidate category as General, ST and SC for 2009 & 2014

fig,ax = plt.subplots(figsize =(8,6))
sns.countplot(x = new_df_candidate2009["Candidate Category"] )
ax.bar_label(ax.containers[-1], fmt = "count: \n%.2f", label_type = "center")
plt.title("seats won based on candidate category")
plt.show()


fig, ax = plt.subplots(figsize = (8,6))
sns.countplot(x = new_df_candidate2014["Candidate Category"])
ax.bar_label(ax.containers[-1], fmt="count: \n%.1f", label_type = "center")
plt.title("seats won based on candidate category")
plt.show()
```

```python
# 5. Plot the age distribution of winners of both 2014 and 2009 elections

fig, ax = plt.subplots(figsize = (8,6))
sns.histplot(x = new_df_candidate2014["Candidate Age"], bins = 10)
ax.bar_label(ax.containers[-1], fmt = "count: \n%.2f", label_type = "center")
plt.title("age distribution of winners of 2014 ")
plt.show()


fig,ax = plt.subplots(figsize =(8,6))
sns.histplot(x = new_df_candidate2009["Candidate Age"], bins =10)
ax.bar_label(ax.containers[-1], fmt = "count: \n%.2f")
plt.title("age distribution of winners of 2009 ")
plt.show()
```

```python
# 6. Separately plot age distribution of NDA & UPA candidates


fig,ax = plt.subplots(figsize=(8,6))

sns.histplot(x = new_df_candidate2009[new_df_candidate2009["Alliences"] == "NDA"]["Candidate Age"], bins = 10)
ax.bar_label(ax.containers[-1], fmt = "count: \n%.2f",label_type = "center")
plt.title("Separately plot age distribution of NDA candidates in 2009")
plt.show()


fig,ax = plt.subplots(figsize=(8,6))
sns.histplot(x = new_df_candidate2009[new_df_candidate2009["Alliences"] == "UPA"]["Candidate Age"], bins = 10)
ax.bar_label(ax.containers[-1], fmt = "count: \n%.2f",label_type = "center")
plt.title("Separately plot age distribution of UPA candidates in 2009")
plt.show()


fig,ax = plt.subplots(figsize=(8,6))
sns.histplot(x = new_df_candidate2014[new_df_candidate2014["Alliences"]=="UPA"]["Candidate Age"], bins = 10)
ax.bar_label(ax.containers[-1], fmt = "count: \n%.2f", label_type = "center")
plt.title("Separately plot age distribution of UPA candidates in 2014")
plt.show()


fig,ax = plt.subplots(figsize=(8,6))
sns.histplot(x = new_df_candidate2014[new_df_candidate2014["Alliences"]=="NDA"]["Candidate Age"], bins = 10)
ax.bar_label(ax.containers[-1], fmt = "count: \n%.2f", label_type = "center")
plt.title("Separately plot age distribution of NDA candidates in 2014")
plt.show()

```

```python
# 7. Plot Gender distributions of 2009 & 2014 elections


fig,ax = plt.subplots(figsize=(8,6))
sns.countplot(x = new_df_candidate2014["Candidate Sex"])
ax.bar_label(ax.containers[-1], fmt = "count: \n%.2f", label_type = "center")
plt.title("Gender distributions of 2014")
plt.show()


fig,ax = plt.subplots(figsize=(8,6))
sns.countplot(x = new_df_candidate2009["Candidate Sex"])
ax.bar_label(ax.containers[-1], fmt = "count: \n%.2f", label_type = "center")
plt.title("Gender distributions of 2009")
plt.show()

```

```python
# 8.Plot gender distribution of NDA and UPA separately for 2009 elections

fig, ax= plt.subplots(figsize = (8,6))
sns.countplot(x = new_df_candidate2009[new_df_candidate2009["Alliences"] == "NDA"]["Candidate Sex"])
ax.bar_label(ax.containers[-1], fmt = "count: \n%.2f", label_type = "center")
plt.title("gender distribution of NDA in 2009")
plt.show()

fig, ax= plt.subplots(figsize = (8,6))
sns.countplot(x = new_df_candidate2009[new_df_candidate2009["Alliences"] == "UPA"]["Candidate Sex"])
ax.bar_label(ax.containers[-1], fmt = "count: \n%.2f", label_type = "center")
plt.title("gender distribution of UDA in 2009")
plt.show()

```
