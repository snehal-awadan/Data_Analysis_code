```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
election2009 = pd.read_csv("LS2009Electors.csv")
candidate2014 = pd.read_csv("LS2014Candidate.csv")
candidate2009 = pd.read_csv("LS2009Candidate.csv")
election2014 = pd.read_csv("LS2014Electors.csv")

candidate2014
```

```python
upa = ['INC','NCP', 'RJD', 'DMK', 'IUML','JMM','JD(s)','KC(M)','RLD','RSP','CMP(J)','KC(J)','PPI','MD']
nda = ['BJP','SS', 'LJP', 'SAD', 'RLSP','AD','PMK','NPP','AINRC','NPF','RPI(A)','BPF','JD(U)','SDF','NDPP','MNF','RIDALOS','KMDK','IJK','PNK','JSP','GJM','MGP','GFP','GVP','AJSU','IPFT','MPP','KPP','JKPC','KC(T)','BDJS','AGP','JSS','PPA','UDP','HSPDP','PSP','JRS','KVC','PNP','SBSP','KC(N)','PDF','MDPF']
other = ['YSRCP','AAAP', 'IND', 'AIUDF', 'BLSP', 'JKPDP', 'JD(S)', 'INLD', 'CPI', 'AIMIM','KEC(M)','SWP', 'NPEP', 'JKN', 'AIFB', 'MUL', 'AUDF', 'BOPF', 'BVA', 'HJCBL','JVM','MDMK']
```

```python
# add into df

candidate2009["allience"] = candidate2009["Party Abbreviation"].apply(lambda x: "upa" if x in upa else "nda" if x in nda else "other")

candidate2014["allience"] = candidate2014["Party Abbreviation"].apply(lambda x: "upa" if x in upa else "nda" if x in nda else "other")
candidate2014
```

```python
df_wining_2009 = candidate2009[candidate2009["Position"] == 1]
df_wining_2014 = candidate2014[candidate2014["Position"] == 1]

print(df_wining_2009)
print(df_wining_2014)
```

```python
# distribution for wiining seats

fig,ax = plt.subplots(figsize = (8,6))
sns.countplot(x = df_wining_2009["allience"])
ax.bar_label(ax.containers[-1], fmt = "count: \n%.2f", label_type = "center")
plt.title("distribution for winning seats for 2009")
plt.show()

fig,ax = plt.subplots(figsize = (8,9))
sns.countplot(x = df_wining_2014["allience"])
plt.title("distribution for winning seats for 2014")
ax.bar_label(ax.containers[-1], fmt ="count:\n %.2f", label_type = "center")
plt.show()
```

```python
# Plot comparatively seats won based on candidate category as General, ST and SC for 2009 & 2014

fig,ax = plt.subplots(figsize = (8,6))
sns.countplot(x = df_wining_2009["Candidate Category"])
plt.title("seats won on candidate by category in 2014")
ax.bar_label(ax.containers[-1], fmt = "count: \n %.2f", label_type = "center")
plt.show()


fig,ax = plt.subplots(figsize = (8,6))
sns.countplot(x = df_wining_2014["Candidate Category"])
plt.title("seats won on candidate by category in 2014")
ax.bar_label(ax.containers[-1], fmt = "count: \n %.2f", label_type = "center")
plt.show()
```

```python
# 5. Plot the age distribution of winners of both 2014 and 2009 elections

fig,ax = plt.subplots(figsize = (8,6))
sns.histplot(x = df_wining_2009["Candidate Age"], bins = 10)
ax.bar_label(ax.containers[-1], fmt = "count: \n % .2f", label_type = "center")
plt.title("age distribution of winning for 2009")
plt.show()

fig,ax = plt.subplots(figsize = (8,6))
sns.histplot(x = df_wining_2014["Candidate Age"], bins = 10)
ax.bar_label(ax.containers[-1], fmt = "count: \n %.2f", label_type = "center")
plt.title("age distribution of winning for 2024")
plt.show()
```
