```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df
```

```python
# calculate correlation factor;
corr_fact = df["reading score"].corr(df["writing score"])
corr_fact
```

```python
# scatter plot
plt.figure(figsize = (12,8))
sns.scatterplot(data = df, x = df["reading score"], y = df["writing score"])
plt.show()
```
