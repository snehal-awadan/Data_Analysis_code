```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
df
```

```python
plt.figure(figsize= (12,8))
sns.scatterplot(data = df, x = "sales", y ="salary", hue = "work experience",palette = "plasma", style = "level of education")
plt.show()
```

```python
sns.histplot(x = "salary",data = df, bins=10)
plt.show()

sns.histplot(x = "sales",data = df, bins=10)
plt.show()

sns.histplot(x = "division",data = df, bins=10)
plt.show()

sns.histplot(x = "level of education",data = df, bins=10)
plt.show()

sns.histplot(x = "training level",data = df, bins=10)
plt.show()
```

```python
# boxplot

plt.figure(figsize = (12,6))
sns.boxplot(data = df, x ="sales", y = "division")
plt.xticks(rotation = 55) # for labeling rotation;
plt.show()
```
